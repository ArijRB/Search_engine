### Parsing des requêtes et jugements de pertinence

```python
fname_query = "../data/cacm/cacm.qry"
fname_rel = "../data/cacm/cacm.rel"
qParser = QueryParser()
qParser.buildQueriesCollection(fname_query)
qParser.buildJudgementsCollection(fname_rel)
```

### Création des modèles et mesures

```python
model1 = Vectoriel(indexer, Weighter2(indexer))
model2 = ModeleLangue(indexer)
model3 = OkapiBM25(indexer)
```

```python
k = 10
beta = 1
mesure1 = Precision(k)
mesure2 = Rappel(k)
mesure3 = F_mesure(k,beta)
mesure4 = AvgPrecision()
mesure5 = NDCG(k)
mesure6 = ReciprocalRank()
```

```python
id_to_model = {0: "modèle vectoriel", 1: "modèle de langue", 2: "OkapiBM25"}
id_to_measure = {0: "précision au rang " + str(k), 1:"rappel au rang " + str(k), \
                 2:"f-mesure au rang " + str(k) + " avec pondération " + str(beta), \
                 3:"précision moyenne", 4:"NDCG au rang " + str(k), 5:"Reciprocal Rank"}
```

### Précision au rang k du modèle vectoriel 

```python
evalModel1 = EvalIRModel(model1, mesure1)
mean_score, std_score, scores = evalModel1.eval(qParser, verbose=True)
print("\n> Moyenne des scores : {} - Écart-type des scores : {}".format(mean_score, std_score))
```

### Évaluation de tous les modèles avec chaque mesure

```python
models = [model1, model2, model3]
measures = [mesure1, mesure2, mesure3, mesure4, mesure5, mesure6]
```

```python
for i in range(len(models)):
    model = models[i]
    for j in range(len(measures)):
        measure = measures[j]
        evalModel = EvalIRModel(model, measure)
        mean, std, _ = evalModel.eval(qParser)
        print("--> {} - {}".format(id_to_model[i], id_to_measure[j]))
        print("\tMoyenne : {} - Écart-type : {}".format(mean,std))
```

### Test de significativité entre le modèle vectoriel et le modèle de langue, avec la précision comme mesure

```python
evalModel1_1 = EvalIRModel(model1, mesure1)
_, _, scores1 = evalModel1_1.eval(qParser)
evalModel2_1 = EvalIRModel(model2, mesure1)
_, _, scores2 = evalModel2_1.eval(qParser)
diff, stat = evalModel1_1.significantly_different(scores1, scores2)
if diff:
    print("Significativement différents (statistique de test : {})".format(stat))
else:
    print("Non significativement différents (statistique de test : {})".format(stat))
```
